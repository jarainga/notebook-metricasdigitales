```python
import pandas as pd
pd.options.display.max_colwidth = None.
```

```python
base = pd.read_excel(r"D:\Users\JONATAN\Downloads\Historicos.xlsx", sheet_name="base")
#base[['vistas', 'likes']] = base[['vistas', 'likes']].astype(str).apply(lambda x: x.str.replace(r'(\d)(?=(\d{3})+$)', r'\1,', regex=True))
base["Año"] = base["Fecha_Estandar"].dt.year
base["Mes"] = base["Fecha_Estandar"].dt.month
base = base.sort_values(by="vistas", ascending=False)
base = base.reset_index(drop=True)
base.head(20)
```

```python
# Promedio de metricas por categoria_progrma
promedios_por_anio = base.groupby(["Año", "Evento_Deportivo"])[["vistas", "likes"]].mean().round(0).astype(int).reset_index()
promedios_por_anio = promedios_por_anio.sort_values(by=["Año", "vistas"], ascending=[True, False])

print(promedios_por_anio)
```
